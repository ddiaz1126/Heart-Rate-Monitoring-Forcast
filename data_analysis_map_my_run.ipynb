```python
# Import Libraries
import pandas as pd
```

```python
import xml.etree.ElementTree as ET
import json

# Parse the TCX file
tree = ET.parse('/Users/daviddiaz/Desktop/Main/Heart-Rate-Monitoring-Forecast/MapMyRunData/exercises/5.00mi Run (3).tcx')  # Replace with the actual file path
root = tree.getroot()

# Define the namespaces used in the TCX file (necessary to access elements)
namespaces = {
    'ns': 'http://www.garmin.com/xmlschemas/TrainingCenterDatabase/v2'
}

# Initialize a list to store the extracted trackpoint data
trackpoints_data = []

# Iterate over the trackpoints in the XML
for trackpoint in root.findall('.//ns:Activities/ns:Activity/ns:Lap/ns:Track/ns:Trackpoint', namespaces):
    trackpoint_info = {}
    
    # Extract relevant information from each trackpoint
    trackpoint_info['time'] = trackpoint.find('ns:Time', namespaces).text
    trackpoint_info['heart_rate'] = trackpoint.find('.//ns:HeartRateBpm/ns:Value', namespaces).text if trackpoint.find('.//ns:HeartRateBpm/ns:Value', namespaces) is not None else None
    trackpoint_info['latitude'] = trackpoint.find('.//ns:Position/ns:LatitudeDegrees', namespaces).text if trackpoint.find('.//ns:Position/ns:LatitudeDegrees', namespaces) is not None else None
    trackpoint_info['longitude'] = trackpoint.find('.//ns:Position/ns:LongitudeDegrees', namespaces).text if trackpoint.find('.//ns:Position/ns:LongitudeDegrees', namespaces) is not None else None

    # Append each trackpoint's data to the list
    trackpoints_data.append(trackpoint_info)

# Convert the list of trackpoint data to JSON format
json_data = json.dumps(trackpoints_data, indent=4)
```

```python
# Save the JSON data to a file
with open('output_data.json', 'w') as json_file:
    json_file.write(json_data)
```
